```python
from bs4 import BeautifulSoup
import requests
import time
import random
from random import randint
import re
```

```python
url='https://www.netbet.fr/football/france/96-ligue-1-conforama'
r = requests.get(url)
print(url, r.status_code)
```

```python
soup = BeautifulSoup(r.content,'html')
soup
```

```python
for p in soup.find_all('a'):
    print (p.text.strip())
```

```python
equipes1 = []
equipes2 = []
for elem in soup.find_all('a', attrs={"class" :"uk-margin-right uk-text-truncate"}):
    equipes1.append(elem.text.strip().split(' / ')[0])
    equipes2.append(elem.text.strip().split(' / ')[1])
```

```python
print(equipes1)
print(equipes2)
```

```python
liste = soup.find_all('a',attrs={"class" :"nb-load uk-flex"})
parie =[]
n=0

for i in range(0,len(liste),6):
     parie.append(liste[i:i+6])
        
```

```python
parie = parie[:len(equipes1)]
```

```python
nul = []
for elem in parie:
    #print(elem)
    nul.append(elem[1].text)
    
print(nul)
```

```python
domicile = []
exterieur = []
for elem in parie:
    domicile.append(elem[0].text)
    exterieur.append(elem[2].text)
```

```python
print(domicile)
print(exterieur)
```

```python
domicile1 = dict(zip(equipes1, domicile))
domicile1
```

```python
exterieur1 = dict(zip(equipes2, exterieur))
exterieur1
```

```python
ligue_1 = []
for i in range(len(equipes1)):
    ligue_1.append({equipes1[i]:domicile[i], "Null":nul[i], equipes2[i]:exterieur[i]})
    
```

```python
# Save a dictionary into a pickle file.
import pickle

pickle.dump(ligue_1, open( "NetBet.p", "wb" ) )
```
